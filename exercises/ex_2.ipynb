```
function is_prime(n: integer) -> boolean:
    if n <= 1 then:
        return False
    else if n = 2 or n = 3:
        return True
    else:
        i <- 2
        while i*i <= n do:
            if n % i == 0 then:
                return False
            end if

        end while

end function
```

````python
def is_prime(n:int = 0) -> bool:
    if n <= 1:
        return False
    elif n == 2 or n == 3:
        return True
    else:
        i = 2
        while i*i<=n:
            if n%i==0:
                return False
            i += 1
    return True
````

````python
# Esempio: Stampa i numeri primi da 0 a 100
for i in range(100):
    if is_prime(i):
        print(i)
````

### Ex 2: Squared Ordered List

Pseudocodice

```
function squared_list(n: integer[]) -> integer[]:
    n <- len(A)
    i <- 0
    j <- n-1
    pos <- n-1
    result = [0]*n
    while i<j do:
        left_sq <- A[i]*A[i]
        right_sq <- A[j]*A[j]
        if left_sq < right_sq then:
            result[pos] <- right_sq
            j <- j-1
        else:
            result[pos] <- left_sq
            i <- i + 1
        end if
        pos <- pos -1
    end while
    return result
end function
```

````python
def squared_list(A: list=[]) -> list:
    n = len(A)
    result = [0] * n
    i, j = 0, n - 1
    pos = n - 1
    while i <= j:
        left_sq = A[i] * A[i]
        right_sq = A[j] * A[j]
        if left_sq > right_sq:
            result[pos] = left_sq
            i += 1
        else:
            result[pos] = right_sq
            j -= 1
        pos -= 1
    return result
````

````python
# Esempio
a = [-20, -4, -1, 0, 3, 10, 100]
print(a, '->', squared_list(a))
````
