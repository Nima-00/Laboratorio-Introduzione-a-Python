### Ex 1: Verificare se un numero è primo

Pseudocodice

```
function is_prime(n: integer) -> boolean:
    if n <= 1 then:
        return False
    else if n = 2 or n = 3:
        return True
    else:
        i <- 2
        while i*i <= n do:
            if n % i == 0 then:
                return False
            end if
            i <- i + 1
        end while
    return True

```

````python
# Implementazione in Python

def is_prime(n:int = 0) -> bool:
    if n <= 1:
        return False
    elif n == 2 or n == 3:
        return True
    else:
        i = 2
        while i*i<=n:
            if n%i==0:
                return False
            i += 1
    return True
````

````python
# Esempio: Stampa i numeri primi da 0 a 100
for i in range(100):
    if is_prime(i):
        print(i)
````

### Ex 2: Squared Ordered List

Pseudocodice

```
function squared_list(n: integer[]) -> integer[]:
    n <- len(A)
    i <- 0
    j <- n-1
    pos <- n-1
    result = [0]*n
    while i<j do:
        left_sq <- A[i]*A[i]
        right_sq <- A[j]*A[j]
        if left_sq < right_sq then:
            result[pos] <- right_sq
            j <- j-1
        else:
            result[pos] <- left_sq
            i <- i + 1
        end if
        pos <- pos -1
    end while
    return result
end function
```

````python
# Implementazione in Python

def squared_list(A: list=[]) -> list:
    n = len(A)
    result = [0] * n
    i, j = 0, n - 1
    pos = n - 1
    while i <= j:
        left_sq = A[i] * A[i]
        right_sq = A[j] * A[j]
        if left_sq > right_sq:
            result[pos] = left_sq
            i += 1
        else:
            result[pos] = right_sq
            j -= 1
        pos -= 1
    return result
````

````python
# Esempio
a = [-20, -4, -1, 0, 3, 10, 100]
print(a, '->', squared_list(a))
````

````python
# Implementazione in Python

def is_palindrome(a: str) -> bool:
    n = len(a)
    for i in range(len(a)//2):
        if a[i] != a[-1-i]:
            return False
    return True
````

### Ex 4: Controllare se una password soddisfa i rewuisiti minimi

````python
def validate_password(password: str) -> bool:

    if len(password)<12:        # Controlla lunghezza
        print("WARNING: Password troppo corta, almeno 12 caratteri")
        return False
    
    has_lower = False           # Controlla lettera minuscola
    for char in password:
        if char.islower():
            has_lower = True
            break
    if not has_lower:
        print("WARNING: Almeno una lettera minuscola")
        return False
    
    has_upper = False           # Controlla lettera maiuscola
    for char in password:
        if char.isupper():
            has_upper = True
            break
    if not has_upper:
        print("WARNING: Almeno una lettera maiuscola")
        return False
    
    has_digit = False           # Controlla numero
    for char in password:
        if char.isdigit():
            has_digit = True
            break
    if not has_digit:
        print("WARNING: Almeno un numero")
        return False
    
    special_chars = "!@#$%^&*"  # Controlla carattere speciale
    has_special = False
    for char in password:
        if char in special_chars:
            has_special = True
            break
    if not has_special:
        print("WARNING: Almeno un carattere speciale")
        return False
    
    return True
````

````python
# Esempio

password = "ndLNbdb24cA!1"
validate_password(password)
````

### Ex 5: Sistema di criptazione di cesare

````python
def caesar_encrypt(text, key):
    """Caesar cipher encoder"""
    result = ""
    
    for char in text:
        # Check if character is uppercase letter
        if char.isupper():
            # Shift character and wrap around if necessary
            result += chr((ord(char) - ord('A') + key) % 26 + ord('A'))
        
        # Check if character is lowercase letter
        elif char.islower():
            # Shift character and wrap around if necessary
            result += chr((ord(char) - ord('a') + key) % 26 + ord('a'))
        
        # If not a letter, keep it unchanged (spaces, punctuation, etc.)
        else:
            result += char
    
    return result
````

````python
testo_originale = "Ciao come va?"
chiave = 10
testo_criptato = caesar_encrypt(testo_originale, chiave)
testo_decriptato = caesar_encrypt(testo_criptato, -chiave)
print(testo_originale)
print(testo_criptato)
print(testo_decriptato)
````

### Ex 6: Contare la frequenza dei caratteri in un testo

````python
def text_frequency(text: str = "") -> dict:
    alphabet = "abcdefghijklmnopqrstuvwxyz0123456789!?.,;:- "
    frequency = {}
    
    # Inizializziamo il conteggio di ogni carattere a 0
    for char in alphabet:
        frequency[char] = 0
    frequency["altro"] = 0
    
    # Contiamo la frequenza dei caratteri
    for char in text:
        if char in frequency:
            frequency[char] += 1
        else:
            frequency["altro"] += 1
    
    return frequency
````

````python
import matplotlib.pyplot as plt


def plot_histogram(frequency: dict):

    chars = [char for char in frequency.keys() if char != "altro"]
    counts = [frequency[char] for char in chars]
    chars.append("altro")
    counts.append(frequency["altro"])
    
    # Creazione Immagine
    plt.figure(figsize=(6, 2.5))
    plt.bar(range(len(chars)), counts, color='steelblue')
    
    plt.xticks(range(len(chars)), chars, fontsize=9)
    plt.xlabel('Caratteri')
    plt.ylabel('Conteggio')
    plt.title('Frequenza dei Caratteri')
    plt.grid(axis='y', alpha=0.3)
    plt.tight_layout()
    plt.show()
````

````python
# Example usage
testo = "Hello, world!"
freq = text_frequency(testo)
plot_histogram(freq)
````
