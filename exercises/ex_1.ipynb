### Ex 1: Conversione Decimale Binario

Pseudocodice

```
function dec_to_bin(n: integer) -> integer
    r <- 0
    i <- 0
    while n > 0:
        r <- r + (n % 2)*10**i
        n <- n//2
        i <- i + 1
    end while
    return r
end function
```

````python
# Implementazione in Python

def dec_to_bin(n:int=0) -> int:
    r = 0
    i = 0
    while n > 0:
        r += (n % 2)*10**i
        n = n//2
        i += 1
    return r
````

````python
# Esempio
a, b, c = 1378, 512, 673
print("Decimale \t| Binario")
print(f"{a} \t\t| {dec_to_bin(a)}")
print(f"{b} \t\t| {dec_to_bin(b)}") 
print(f"{c} \t\t| {dec_to_bin(c)}") 
````

### Ex 3: Massimo Comune Divisore

Pseudocodice

```
function alg_euclide(a: integer, b: integer) -> integer:
    if b < a:
        c <- a
        a <- b
        b <- c
    end if
    while b != 0:
        tmp <- a
        a <- b
        b <- tmp % b
    end while
    return a
end function
```

````python
# Implementazione in Python

def alg_euclide(a: int, b: int) -> int:
    if b<a:
        c = a
        a = b
        b = c
    while b != 0:
        tmp = a
        a = b
        b = tmp%b
    return a
````

````python
# Esempio
a, b, c, d, e, f = 173, 439, 34, 28, 500, 250
print(f"MCD({a},{b}) = ", alg_euclide(a, b))
print(f"MCD({c},{d}) = ", alg_euclide(c, d))
print(f"MCD({e},{f}) = ", alg_euclide(e, f))
````

````python
# Implementazione in Python

def max_one_seq(n: int) -> int:
    max = 0
    c_max = 0
    while n > 0:
        if n%10==1:
            c_max += 1
            if max<c_max: max=c_max
        else:
            c_max = 0
        n = n//10
    return max
````

````python
# Esempio
a, b, c = 1101001, 1011110, 1010101
print(f"Max # di 1 consecutivi in {a}:\t", max_one_seq(a))
print(f"Max # di 1 consecutivi in {b}:\t", max_one_seq(b))
print(f"Max # di 1 consecutivi in {c}:\t", max_one_seq(c))
````

### Ex 5: numero di numeri con un numero pari di cifre

Pseudocodice

```
function n_cifre_pari(a: integer) -> boolean:
    pari <- True
    counter <- 0
    while a > 0:
        counter = counter + 1
        a <- a // 10
    end while
    if counter%2 != 0:
        pari <- False
    end if
    return pari
end function  

function n_pari_couter(a: integer[]) -> integer:
    n <- 0
    for i in [0 ... len(a)) do:
        if n_cifre_pari(a[i]):
            n <- n + 1
        end if
    end for
    return n
end function
```

````python
# Implementazione in Python

def n_cifre_pari(a: int) -> bool:
    pari = True
    counter = 0
    while a > 0:
        counter = counter + 1
        a = a // 10
    if counter%2!=0: pari = False
    return pari

def n_pari_couter(a: list) -> int:
    n = 0
    for i in range(len(a)):
        if n_cifre_pari(a[i]):
            n = n + 1
    return n
````

````python
# Esempio
a, b, c = [1234, 10, 143], [1002, 32, 14], [1, 132, 432]
print(f"# di numeri pari in {a}:\t", n_pari_couter(a))
print(f"# di numeri pari in {b}:\t", n_pari_couter(b))
print(f"# di numeri pari in {c}:\t", n_pari_couter(c))
````
