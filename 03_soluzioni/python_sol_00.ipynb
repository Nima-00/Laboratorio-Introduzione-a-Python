### Ex 1: Verificare se un numero è primo

Scrivi una funzione in pseudocodice per verificare se un numero è primo, poi implementala con Python, infine stampa i numeri primi da 1 a 100.

Pseudocodice

```
function is_prime(n: integer) -> boolean:
    if n <= 1 then:
        return False
    else if n = 2 or n = 3:
        return True
    else:
        variable i: integer
        i <- 2
        while i*i <= n do:
            if n mod i == 0 then:
                return False
            end if
            i <- i + 1
        end while
    end if
    return True
end function
```

````python
# Implementazione in Python

def is_prime(n:int = 0) -> bool:
    if n <= 1:
        return False
    elif n == 2 or n == 3:
        return True
    else:
        i = 2
        while i*i<=n:
            if n%i==0:
                return False
            i += 1
    return True
````

````python
# Esempio: Stampa i numeri primi da 0 a 100
for i in range(100):
    if is_prime(i):
        print(i)
````

### Ex 2: Lista di quadrati ordinata

Data una lista ordinata calcolare la lista dei quadrati ordinata (es. [-2, 1, 3] -> [1, 4, 9])

Pseudocodice

```
function squared_list(A: integer[], n:integer) -> integer[]:
    variable i: integer
    variable j: integer
    variable pos: integer
    variable result: integer[]
    i <- 0
    j <- n-1
    pos <- n-1
    result <- [0, ..., 0]   # array lungo n
    variable left_sq: integer
    variable right_sq: integer
    while i<j do:
        left_sq <- A[i]*A[i]
        right_sq <- A[j]*A[j]
        if left_sq < right_sq then:
            result[pos] <- right_sq
            j <- j-1
        else:
            result[pos] <- left_sq
            i <- i + 1
        end if
        pos <- pos -1
    end while
    return result
end function
```

````python
# Implementazione in Python

def squared_list(A: list=[]) -> list:
    n = len(A)
    result = [0] * n
    i, j = 0, n - 1
    pos = n - 1
    while i <= j:
        left_sq = A[i] * A[i]
        right_sq = A[j] * A[j]
        if left_sq > right_sq:
            result[pos] = left_sq
            i += 1
        else:
            result[pos] = right_sq
            j -= 1
        pos -= 1
    return result
````

````python
# Esempio
a = [-20, -4, -1, 0, 3, 10, 100]
print(a, '->', squared_list(a))
````

### Ex 3: Controllare se un numero è palindromo

````python
# Implementazione in Python

def is_palindrome(a: str) -> bool:
    n = len(a)
    for i in range(len(a)//2):
        if a[i] != a[-1-i]:
            return False
    return True
````

### Ex 4: Controllare se una password soddisfa i requisiti minimi di sicurezza

I requisiti minimi sono:
- lunghezza di almeno 12 caratteri
- almeno una lettera minuscola
- almeno una lettera maiuscola
- almeno una cifra
- almeno un carattere speciale tra !@#$%^&*

Si possono utilizzare i metodi implementati per le stringhe .islower(), .isupper(), .isdigit().

Per vedere come funzionano utilizzare il comando help() (es. help(str.isdigit))

````python
def validate_password(password: str) -> bool:

    if len(password)<12:        # Controlla lunghezza
        print("WARNING: Password troppo corta, almeno 12 caratteri")
        return False
    
    has_lower = False           # Controlla lettera minuscola
    for char in password:
        if char.islower():
            has_lower = True
            break
    if not has_lower:
        print("WARNING: Almeno una lettera minuscola")
        return False
    
    has_upper = False           # Controlla lettera maiuscola
    for char in password:
        if char.isupper():
            has_upper = True
            break
    if not has_upper:
        print("WARNING: Almeno una lettera maiuscola")
        return False
    
    has_digit = False           # Controlla numero
    for char in password:
        if char.isdigit():
            has_digit = True
            break
    if not has_digit:
        print("WARNING: Almeno un numero")
        return False
    
    special_chars = "!@#$%^&*"  # Controlla carattere speciale
    has_special = False
    for char in password:
        if char in special_chars:
            has_special = True
            break
    if not has_special:
        print("WARNING: Almeno un carattere speciale")
        return False
    
    return True
````

````python
# Esempio

password = "ndLNbdb24cA!1"
validate_password(password)
````
