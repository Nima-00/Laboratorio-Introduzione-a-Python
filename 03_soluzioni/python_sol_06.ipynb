Scrivi una funzione che prende in input un numero intero positivo n e stampa i numeri divisibili per 3 tra 0 e n.

	function stampa_multipli(n: integer):
		for i in [0, n+1) do:
			if i mod 3 == 0 then:
				output(i)

```python
# Implementazione in Python
def stampa_multipli(n: int):
	for i in range(0, n+1):
		if i % 3 == 0:
			print(i)
```

### Ex 2: Bim Bum Bam

Scrivi una funzione che prende in input un intero positivo e stampa i numeri da 0 a n-1 sostituendo i multipli di 21 con "Bim!", i multipli di 3 con "Bum!" e i multipli di 7 con "Bam!"

Esempio: funzione(8) stamperà:

	"Bim!"
	1
	2
	"Bum!"
	4
	5
	"Bum!"
	"Bam!"

	function bimbumbam(n: integer):
		for i in [0, n-1) do:
			if i mod 21 == 0 then:
				output("Bim!")
			else if i mod 3 == 0 then:
				output("Bum!")
			else if i mod 7 == 0 then:
				output("Bam!")
			else:
				output(i)
			end if
		end for
	end function

```python
def bimbumbam(n: int):
	for i in range(n):
		if i % 21 == 0:
			print("Bim!")
		elif i % 3 == 0:
			print("Bum")
		elif i % 7 == 0:
			print("Bam")
		else:
			print(i)
```

```python
bimbumbam(24)
```

### Ex 3:

Scrivi una funzione che prende in input un array e la sua lunghezza e restituisce la somma degli elementi in posizione pari meno la somma degli elementi in posizione dispari.

Esempio: fun_1([1, 2, 3, 4, 5, 6], 6) restituisce (1+3+5) - (2+4+6) = -3

	function fun_1(A: integer[], n: integer) -> integer:
		variable s_p: integer
		variable s_d: integer
		s_p <- 0
		s_d <- 0
		for i in [0, n) do
			if i mod 2 == 0 then:
				s_p <- s_p + A[i]
			else:
				s_d <- s_d + A[i]
			end if
		end for
		return s_p - s_d
	end function

```python
def fun_1(A: list, n: int) -> int:
	s_p = 0
	s_d = 0
	for i in range(n):
		if i % 2 == 0:
			s_p = s_p + A[i]
		else:
			s_d = s_d + A[i]
	return s_p - s_d
```

```python
A = [1, 2, 3, 4, 5, 6]
n = len(A)
b = fun_1(A, n)
print(b)
```

### Ex 4:

Scrivi una funzione che prende in input un array, la sua lunghezza, e un valore. Ritorna il numero di elementi maggiori dal valore dato in input.

Esempio: fun_2([8, 13, 2, 3, 7], 5, 4) ritorna 3

	function fun_2(A: integer[], n: integer, val: integer) -> integer:
		variable k: integer
		k <- 0
		for i in [0, n) do:
			if A[i] > val then:
				k <- k + 1
			end if
		end for
		return k
	end function

```python
def fun_2(A: list, n: int, val: int) -> int:
    k = 0
    for i in range(n):
        if A[i] > val:
            k = k + 1
    return k
```

```python
a = fun_2([8, 13, 2, 3, 7], 5, 4)
print(a)
```

### Ex 5:

Scrivi una funzione che prende in input un numero e continua a sommare le sue cifre fino ad ottenerne una.

Esempio: fun(289) ritorna 1 perchè 289 -> 19 -> 10 -> 1

	function somma_cifre(n: integer) -> integer:
		variable s: integer
		while n // 10 > 0:
			s <- 0
			while n > 0 do:
				s <- s + n mod 10
				n <- n // 10
			end while
			n <- s
		end while
		return n
	end function

```python
# Implementazione in Python
def somma_cifre(n: int) -> int:
    while n // 10 > 0:
        s = 0
        while n > 0:
            s = s + n % 10
            n = n // 10
        n = s
    return n
```

```python
# Esempio
a = somma_cifre(289)
print(a)
```
