### Ex 1: Stampare gli elementi di un array

    function stampa_array(A: integer[], n: integer):
        for i in [0, n) do:
            output(A[i])
        end for
    end function

```python
def stampa_array(A: list) -> None:
    n = len(A)
    for i in range(n):
        print(A[i])
```

```python
A = [2, 3, 2, 5]
stampa_array(A)
```

### Ex 2: Stampare gli elementi pari 

    function stampa_array_pari(A: integer[], n: integer):
        for i in [0, n) do:
            if A[i] mod 2 = 0 then:
                output(A[i])
            end if
        end for
    end function

```python
def stampa_array_pari(A: list) -> None:
    n = len(A)
    for i in range(n):
        if A[i] % 2 == 0:
            print(A[i])
```

```python
B = [0, 1, 2, 3, 4, 5, 6]
stampa_array_pari(B)
```

### Ex 3: Stampa elementi in posizione pari o dispari in base a ciò che è richiesto

    function stampa_elementi_array(A: integer[], n: integer,  pari: bool):
        variable m: integer
        if pari then:
            m <- 0
        else:
            m <- 1
        end if
        for i in [0, n) do:
            if i mod 2 = m then:
                output(A[i])
            end if
        end for
    end function

```python
def stampa_elementi_array(A: list, pari: bool) -> None:
    m = 1
    if pari:
        m = 0
    for i in range(len(A)):
        if i % 2 == m:
            print(A[i])
```

```python
B = [0, 1, 2, 3, 4, 5]
stampa_elementi_array(B, pari=False)
```

### Ex 4: Somma array

    function somma_array(A: integer[], n: integer) -> integer:
        variable s: integer
        s <- 0
        for i in [0, n) do:
            s <- s + A[i]
        end for
        return s
    end function

```python
def somma_array(A: list) -> int:
    s = 0
    for i in range(len(A)):
        s = s + A[i]
    return s
```

```python
B = [1, -1, 1, -1, 1, -1]
somma_array(B)
```

### Ex 5: inverti array

    function inverti_array(A: integer[], n: integer) -> integer[]:
        variable tmp: integer
        for i in [0, n//2) do:
            tmp <- A[i]
            A[i] <- A[n-1-i]
            A[n-1-i] <- A[i]
        return A
    end function

```python
def inverti_array(A: list):
    n = len(A)
    for i in range(n//2):
        tmp = A[i]
        A[i] = A[n-1-i]
        A[n-1-i] = tmp
    return A
```

```python
B = [0, 1, 2, 3, 4, 5]
inverti_array(B)
```

### Ex 6: Ritorna array con elementi in posizione pari

    function elementi_pari(A: integer[], n: integer) -> integer[]:
        variable B: integer[]
        for i in [0, n) do:
            if i mod 2 = 0 then:
                B.append(A[i])
            end if
        end for
        return B
    end function

```python
def elementi_pari(A: list) -> list:
    n = len(A)
    B = []
    for i in range(n):
        if i % 2 == 0:
            B.append(A[i])
    return B
```

```python
B = [0, 1, 2, 3, 4]

elementi_pari(B)
```

### Ex 7: Stampa le singole cifre di un numero

    function stampa_singole_cifre(n: integer):
        while n > 0 do:
            output(n mod 10)
            n <- n//10
        end while
    end function

```python
def stampa_singole_cifre(n: int) -> None:
    while n > 0:
        print(n%10)
        n = n//10
```

```python
stampa_singole_cifre(12345)
```

### Ex 9: Numero cifre pari

    function n_cifre_pari(n: integer) -> bool:
        variable counter: integer
        counter <- 0
        while n > 0 do:
            counter <- counter +1
            n <- n // 10
        end while
        if counter mod 2 = 0 then:
            return True
        else:
            return False
        end if
    end function

```python
def n_cifre_pari(n: int) -> bool:
    counter = 0
    while n > 0:
        counter = counter + 1
        n = n // 10
    if counter % 2 == 0:
        return True
    else:
        return False
```

```python
n_cifre_pari(12324)
```

### Ex 10: Inverti l'ordine delle cifre

    function inverti_cifre(n: integer) -> integer:
        variable r: integer
        r <- 0
        while n > 0 do:
            r <- (n mod 10) + r*10
            n <- n//10
        end while
        return r
    end function

```python
def inverti_cifre(n: int) -> int:
    r = 0
    while n > 0:
        r = r*10 + n%10
        n = n//10
    return r
```

```python
inverti_cifre(12345)
```
