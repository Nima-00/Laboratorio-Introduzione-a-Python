### Ex 2: Successione di Fibonacci

La successione di Fibonacci inizia con i termini F_0 = 0 e F_1 = 1. Da qui in poi ogni termine è calcolato come la somma dei due preccedenti. Scrivi una funzione che prende in input un intero n e restituisca come output F_n, l'ennesimo elemento della successione di Fibonacci.

<!-- Pseudocodice -->

    function fibonacci(n: integer) -> integer:
        variable f_0: integer
        variable f_1: integer
        variable tmp: integer
        f_0 <- 0
        f_1 <- 1
        for i in [1, n) do:
            tmp <- f_0 + f_1
            f_0 <- f_1
            f_1 <- tmp
        end for
        if n == 0 then:
            return f_0
        else:
            return f_1
        end if
    end function

```python
# Implementazione in Python
def fibonacci(n):
    f_0 = 0
    f_1 = 1
    for i in range(1, n):
        f_n = f_0 + f_1
        f_0 = f_1
        f_1 = f_n
    if n == 0:
        return f_0
    else:
        return f_1
```

```python
for i in range(10):
    print(fibonacci(i))
```

### Ex 3: Somma elementi in posizione pari e somma elementi in posizione dispari

Scrivi una funzione che prende in input un array e la sua lunghezza e restituisca come output la somma degli elementi in posiione pari e la somma degli elementi in posizione dispari.

<!-- Pseudocodice -->

    function somma_array(A: integer[], n: integer) -> integer[]:
        variable S: integer[2]
        S <- [0, 0]
        for i in [0, n) do:
            if i mod 2 == 0 then:
                S[0] <- S[0] + A[i]

            end if

```python
def somma_array(A: list, n: int) -> int:
    S = [0, 0]
    for i in range(n):
        if i % 2 == 0:
            S[0] += A[i]
        else:
            S[1] += A[i]
    return S
```

```python
A = [1, 2, 3, 4, 5]
n = len(A)
B = somma_array(A, n)
print(B)
```

### Ex 4: Prodotto delle cifre pari

Scrivi una funzione che prende in input un numero intero e stampa e restituisce il prodotto delle cifre pari

<!-- Pseudocodice -->

    function prodotto_cifre_pari(n: integer) -> integer:
        variable p: integer
        p <- 1
        while n > 0 do:
            cifra = n % 10
            if cifra mod 2 == 0 then:
                p <- p x cifra
            end if
            n <- n / 10
        end while
    end function

```python
def prodotto_cifre_pari(n: int) -> int:
    p = 1
    while n > 0:
        cifra = n % 10
        if cifra % 2 == 0:
            p = p*cifra
        n = n // 10
    if p == 1:
        return 0
    return p
```

```python
prodotto_cifre_pari(135792)
```

### Ex 5: Trovare i tre numeri più piccoli in un array

Scrivi una funzione che prenda in input un array e la sua lunghezza e restituisca i tre numeri più piccoli.

<!-- Pseudocodice -->

    function ordina_tre_valori(a: integer, b: integer, c: integer) -> integer[]:
        if a <= b then:
            if b <= c then:
                return [a, b, c]
            else if a <= c then:
                return [a, c, b]
            else:
                return [c, a, b]
            end if
        else:
            if a <= c then:
                return [b, a, c]
            else if b <= c then:
                return [b, c, a]
            else:
                return [c, b, a]
            end if
        end if
    end function

    function tre_min(A: integer[], n: integer) -> integer[]:
        variable M: integer[]
        M <- ordina_tre_valori(A[0], A[1], A[2])
        for i in [0, n) do:
            if A[i] < M[0] then:
                M[2] <- M[1]
                M[1] <- M[0]
                M[0] <- A[i]
            else if A[i] < M[1] then:
                M[2] <- M[1]
                M[1] <- A[i]
            else if A[i] < M[2] then:
                M[2] <- A[i]
            end if
        end for
        return M
    end function

```python
def order_three_values(a: int, b: int, c: int) -> list:
    if a <= b:
        if b <= c:
            return [a, b, c]
        elif a <= c:
            return [a, c, b]
        else:
            return [c, a, b]
    else:
        if a <= c:
            return [b, a, c]
        elif b <= c:
            return [b, c, a]
        else:
            return [c, b, a]


def tre_min(A: list, n: int) -> list:
    M = order_three_values(*A[:3])
    for i in range(n):
        if A[i] < M[0]:
            M[2] = M[1]
            M[1] = M[0]
            M[0] = A[i]
        elif A[i] < M[1]:
            M[2] = M[1]
            M[1] = A[i]
        elif A[i] < M[2]:
            M[2] = A[i]
    return M
```

```python
A = [144, 144, 144, 144, 1, 0, 2, 3]
n = len(A)
tre_min(A, n)
```
