### Esercizio 1:

Scrivi una funzione che prende in input un array e la sua lunghezza e ritorna la somma dei suoi elementi.

Esempio: fun_1([5, 2, 8, 3], 4) restituisce 18

	function fun_1(A: integer[], n: integer) -> integer:
		variable k: integer
		k <- 0
		for i in [0, n) do:
			k <- k + A[i]
		end for
		return k
	end function

```python
# Implementazione in Python
def fun_1(A: list, n: int) -> int:
    k = 0
    for i in range(n):
        k = k + A[i]
    return k
```

```python
# Esempio
arr = [1, 2, 3, 4, 5]
n = len(arr)
k = fun_1(arr, n)
print(k)
```

### Esercizio 3:

Scrivi una funzione che prende in input un intero positivo e restituisce il numero di volte che la cifra 4 compare in tale numero.

Esempio: fun_3(148) restituisce 1

	function fun_3(n: integer) -> integer:
		variable k: integer
		k <- 0
		while n > 0 do:
			if n mod 10 == 4 then:
				k <- k +1
			end if
			n <- n // 10
		end while
		return k
	end function

```python
# Implementazione in Python
def fun_3(n: int) -> int:
    k = 0
    while n > 0:
        if n % 10 == 4:
            k = k + 1
        n = n // 10
    return k
```

### Esercizio 5:

Scrivi una funzione che prende in input un array di interi e la sua lunghezza e restituisce la posizione del primo elemento uguale a 0, altrimenti -1.

Esempio: fun_5([1, 2, 3, 0, 4, 5, 0], 7) restituisce 3, mentre fun_5([1, 8, 3, 4], 4) restituisce -1.

	function fun_5(A: integer[], n: integer) -> integer:
		for i in [0, n) do:
			if A[i] == 0:
				return i
			end if
		end for
		return -1
	end function

```python
# Implementazione in Python
def fun_5(A: list, n: int) -> int:
    k = -1
    for i in range(n):
        if A[n - 1 - i] == 0:
            k = n - 1 - i
    return k

# Alternativa
def fun_5_1(A: list, n: int) -> int:
    for i in range(n):
        if A[i] == 0:
            return i
    return -1

def fun_5_2(A: list, n: int) -> int:
    k = -1
    for i in range(n):
        if A[i] == 0:
            k = i
            break
    return k
```

```python
a = [1, 2, 3, 5, 0]
n = len(a)
b = fun_5(a, n)
# b = fun_5_1(a, n)
# b = fun_5_2(a, n)
print(b)
```
