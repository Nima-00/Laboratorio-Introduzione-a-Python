### Ex 1:

Scrivi una funzione che prende in input un intero n e stampa i numeri divisibili per 4 tra n-1 e 0

	function stampa_multipli(n: integer):
		for i in [0, n) then:
			if i mod 4 == 0 then:
				output(i)

```python
def stampa_multipli(n: int):
	for i in range(n):
		if i % 4 == 0:
			print(i)
```

### Ex 3

Scrivi una funzione che prende in input due array e restituisce il prodotto della somma degli elementi di A e la somma degli elementi di B

	function fun_2(A: integer[], B: integer[], n: integer, m: integer) -> integer:
		variable s_A: integer
		variable s_B: integer
		s_A <- 0
		s_B <- 0
		for i in [0, n) do:
			s_A <- s_A + A[i]
		end for
		for j in [0, m) do:
			s_B <- s_B + B[j]
		end for
		return s_A * s_B
	end function

```python
def fun_2(A: list, B: list, n: int, m: int) -> int:
	s_A = 0
	s_B = 0
	for i in range(n):
		s_A = s_A + A[i]
	for j in range(m):
		s_B = s_B + B[j]
	return s_A * s_B
```

```python
v_1 = [8, 13, 2, 7]
v_2 = [1, 2, 3]
n_1 = len(v_1)
n_2 = len(v_2)
r = fun_2(v_1, v_2, n_1, n_2)
print(r)
```

### Ex 5:

Scrivi una funzione che stampa il prodotto del massimo e del minimo

	function fun_3(A: integer[], n: integer):
		variable m: integer
		variable M: integer
		m <- A[0]
		M <- A[0]
		for i in [1, n) do:
			if A[i] < m then:
				m <- A[i]
			else if M < A[i] then:
				M <- A[i]
			end if
		end for
		return m * M
	end function

```python
def fun_3(A: list, n: int) -> int:
    m = A[0]
    M = A[0]
    for i in range(1, n):
        if A[i] < m:
            m = A[i]
        elif M < A[i]:
            M = A[i]
    return m * M
```

```python
v = [8, 13, 2, 7, 5]
n = len(v)
p = fun_3(v, n)
print(p)
```
