    function stampa_array_inv(A: integer[], n: integer):
        variable i: integer
        i <- n-1

### Ex 3: Stampare i primi tre valori più grandi di un array

Note: Il valore minimo è 0

    function tre_max(A: integer[], n: integer) -> integer[]:
        variable M: integer[]
        M <- [-1, -1, -1]
        for i in [0, n) do:
            if M[0] < A[i] then:
                M[2] <- M[1]
                M[1] <- M[0]
                M[0] <- A[i]
            elif M[1] < A[i] then:
                M[2] <- M[1]
                M[1] <- A[i]
            elif M[2] < A[i] then:
                M[2] <- A[i]
            end if
        end for
        return A
    end function

```python
# Implementazione in Python
def tre_max(A: list, n: int) -> list:
    M = [-1, -1, -1]
    for i in range(n):
        if M[0]<A[i]:
            M[2] = M[1]
            M[1] = M[0]
            M[0] = A[i]
        elif M[1]<A[i]:
            M[2] = M[1]
            M[1] = A[i]
        elif M[2]<A[i]:
            M[2] = A[i]
    return M
```

```python
# Esempio
B = [1, 2, 30, 41, 5, 10, 23, 5, 2, 11]
tre_max(B, len(B))
```

### Ex 4: Stabilire se il massimo è almeno due volte più grande di ogni altro numero

Note:
1. La funzione ritorna l'indice dell'elemento se è vero altrimenti -1
2. Il massimo è unico (non ci possono essere due massimi con il medesimo valore)

    function max_idx(A: integer[], n: int) -> int:
        int idx = 0
        for i in [0, n) do:
            if A[idx] < A[i] then:
                idx = i
            end if
        end for
        for i in [0, n) do:
            if idx != i and A[idx] < 2*A[i] then:
                return -1
            end if
        end for
        return idx
    end function

```python
# Implementazione in Python
def max_idx(A: list, n: int) -> int:
    idx = 0
    for i in range(n):
        if A[idx]<A[i]:
            idx = i
    for i in range(n):
        if i != idx and A[idx] < 2*A[i]:
            return -1
    return idx
```

```python
# Esempio
A = [1, 2, 3, 40, 54, 6]
max_idx(A, len(A))
```

### Ex 5: Fare una funzione che stampa i numeri da uno a n sostituendo "Bim!" per i multipli di 15, "Bum! per i multipli di 5 e "Bam!" per i multipli di 3

    function fun(n: integer):
        for i in [1, n+1) do:
            if i mod 15 == 0 then:
                output("Bim!")
            else if i mod 5 == 0 then:
                output("Bum!")
            else if i mod 3 == 0 then:
                output("Bam")
            else:

```python
# Implementazione in Python
def fun(n: int):
    for i in range(1, n+1):
        if i % 15 == 0:
            print("Bim!")
        elif i % 5 == 0:
            print("Bum!")
        elif i % 3 == 0:
            print("Bam!")
        else:
            print(i)
```

```python
# Esempio
fun(10)
```

### Ex 6: Questa funzione prende in input un array in cui in ogni entrata è presente una singola cifra. Sommate uno all'ultima cifra come se fosse un numero naturale e date in output l'array.

    function somma_uno(A: integer[], n: integer) -> integer[]:
        variable i: integer
        i = n-1
        while 0<=i do:
            if A[i] != 9 then:
                A[i] = A[i]+1
                return A
            else
                A[i] = 0
                i = i - 1
            end if
        end while
        A.insert(0, 1)          # Add a 1 before position 0. Ex [2, 2, 2].insert(0, 1) -> [1, 2, 2, 2]
        return A
    end function

```python
def somma_uno(A: list, n: int) -> list:
    i = n-1
    while 0<=i:
        if A[i] != 9:
            A[i] = A[i] + 1
            return A
        else:
            A[i] = 0
            i = i -1
    A.insert(0, 1)
    return(A)
```

```python
A = [1, 0, 9, 9, 3]
somma_uno(A, len(A))
```

### Ex 7: Trovare il primo indice, se esiste, per cui la somma degli elemnti alla sua sinistra e alla sua destra sono uguali. Se esiste restiruire la prima posizione partendo da sinistra altrimenti restituire -1.

    function pivot_idx(A: integer[], n: integer) -> integer:
        variable s: integer
        s <- 0
        for i in [0, n) do:
            s <- s + A[i]
        end for

        variable left_sum: integer
        varibale right_sum: integer
        left_sum <- 0
        right_sum <- s

        for i in [0, n) do:
            right_sum <- right_sum - A[i]
            if left_sum == right_sum then:
                return i
            end if
            left_sum <- left_sum + A[i]
        end for

        return -1

    end function

```python
def pivot_idx(A: list, n: int) -> int:
    s = 0
    for i in range(n):
        s = s + A[i]
    left_sum = 0
    right_sum = s
    for i in range(n):
        right_sum = right_sum - A[i]
        if left_sum == right_sum:
            return i
        left_sum = left_sum + A[i]
    return -1
```

```python
A = [1,7,3,6,5,6]
n = len(A)
pivot_idx(A, n)
```

### Ex 8: Stampa una matrice a serpentina

    function stampa_diagonale(A: integer[][], n: integer, m: integer):
        variable i: integer
        variable j: integer

        for k in [0, n + m - 1) do:
            if k mod 2 == 0 then:
                i <- min(k, n - 1)
                j <- k - i
                while 0 <= i and j < m do:
                    output(A[i][j])
                    i = i + 1
                    j = j - 1
                end while
            else
                j = min(k, m - 1);
                i = k - j;
                while 0 <= j and i < n do:
                    output(A[i][j])
                    i = i - 1
                    j = j + 1
                end while
            end if
        end for
    end function

```python
def stampa_diagonale(A: list, n: int, m: int):
    for k in range(n+m-1):
        if k % 2 == 0:
            i = min(k, n-1)
            j = k - i
            while 0 <= i and j < m:
                print(A[i][j])
                i = i - 1
                j = j + 1
        else:
            j = min(k, m-1)
            i = k - j
            while 0 <= j and i < n:
                print(A[i][j])
                i = i + 1
                j = j - 1
```

```python
mat = [[1,2,3],[4,5,6],[7,8,9]]
stampa_diagonale(mat, len(mat), len(mat[0]))
```
