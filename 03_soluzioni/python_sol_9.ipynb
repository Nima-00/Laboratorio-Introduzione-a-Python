### Esercizio 1:

Scrivi una funzione che prende in input un array e la sua lunghezza e ritorna l'elemento più piccolo.

Esempio: fun_1([5, 2, 8, 3], 4) restituisce 2

	function fun_1(A: integer[], n: integer) -> integer:
		variable m: integer
		m <- A[0]
		for i in [0, n) do:
			if A[i] < m then:
				m <- A[i]
			end if
		end for
		return m
	end function

```python
# Implementazione in Python
def fun_1(A: list, n: int) -> int:
    m = A[0]
    for i in range(n):
        if A[i] < m:
            m = A[i]
    return m
```

```python
# Esempio
arr = [1, 2, 3, 4, 5, 0]
l = len(arr)
k = fun_1(arr, l)
print("L'elemento più piccolo è", k)
```

### Esercizio 2:

Scrivi una funzione che prende in input un array e la sua lunghezza e stampa gli elementi in posizione dispari.

Esempio: fun_2([2, 3, 9, 3], 4) stampa 3 3

	function fun_2(A: list, n: int):
		for i in [0, n) do:
			if i mod 2 == 1 then:
				output(A[i])
			end if
		end for
	end function

```python
# Implementazione in Python
def fun_2(A: list, n: int):
    for i in range(n):
        if i % 2 == 0:
            print(A[i])
```

### Esercizio 3:

Scrivi una funzione che prende in input un array e la sua lunghezza e restituisce la differenza tra la somma degli elementi pari e la somma degli elementi dispari.

Esempio: fun_3([1, 2, 3, 4, 5], 5) restituisce (2+4) - (1+3+5) = -3

	function fun_3(A: integer[], n: integer) -> integer:
		variable s: integer
		s <- 0
		for i in [0, n) do:
			if A[i] mod 2 == 0:
				s <- s + A[i]
			else:
				s <- s - A[i]
			end if
		end for
		return s
	end function

```python
# Implementazione in Python
def fun_3(A: list, n: int) -> int:
    s = 0
    for i in range(n):
        if A[i] % 2 == 0:
            s = s + A[i]
        else:
            s = s - A[i]
    return s
```

```python
# Esempio
arr = [1, 2, 3, 4, 5, 8, 2, 7]
l = len(arr)
fun_3(arr, l)
```

### Esercizio 4:

Scrivi una funzione che prende in input un intero positivo e lo divide per due finchè non diventa dispari.

Esempio: fun_4(12) restituisce 3, perché 12 -> 6 -> 3, fun_4(9) restituisce 9, mentre fun_4(8) restituisce 1. 

	function fun_4(n: integer) -> integer:
		while n mod 2 == 0:
			n <- n // 2
		return n
	end function

```python
# Implementazione in Python
def fun_4(n: int) -> int:
    while n % 2 == 0:
        n = n // 2
    return n
```

```python
# Esempio
a = fun_4(181)
print(a)
```

### Esercizio 5:

Scrivi una funzione che prende in input un intero positivo n e restituisce la sua cifra più grande.

Esempio: fun_5(154) restituisce 5

	function fun_5(n: integer) -> integer:
		variable k_max: integer
		variable k: integer
		k_max <- 0
		while n > 0 do:
			k <- n mod 10
			if k_max < k then:
				k_max <- k
			end if
			n <- n // 10
		end while
		return k
	end function

```python
# Implementazione in Python
def fun_5(n: int) -> int:
    k_max = 0
    while n > 0:
        k = n % 10
        if k_max < k:
            k_max = k
        n = n // 10
    return k_max
```

```python
# Esempio
n = 1832
k = fun_5(n)
print(f'La cifra più grande in {n} è {k}')
```
