### Ex 1: Stampa elementi pari

Scrivi una funzione che prende in input un array e stampa gli elementi pari

	function stampa_elementi_pari(A: integer[], n: integer):
		for i in [0, n) do:
			if A[i] mod 2 == 0 then:
				output(A[i])
			end if
		end for
	end function

```python
# Implementazione in Python
def stampa_elementi_pari(A: list, n: int):
    for i in range(n):
        if A[i] % 2 == 0:
            print(A[i])
```

```python
# Esempio
A = [1, 2, 3, 4, 5, 6]
n = len(A)
stampa_elementi_pari(A, n)
```

### Ex 2

Scrivi una funzione che prende in input due array della stessa lunghezza, restituisce un array in cui ogni elemento contiene il minore delle due entrate corrispondenti.

	function new_array(A: integer[], B: integer[], n: integer) -> integer[]:
		variable C: integer[n]
		for i in [0, n) do:
			if A[i] < B[i] then:
				C[i] <- A[i]
			else:
				C[i] <- B[i]
			end if
		end for
		return C
	end function

```python
# Implementazione in Python
def new_array(A: list, B: list, n: int) -> list:
    C = [0]*n
    for i in range(n):
        if A[i] < B[i]:
            C[i] = A[i]
        else:
            C[i] = B[i]
    return C
```

```python
v_1 = [3, 5, 2, 8]
v_2 = [4, 1, 6, 7]
n = len(v_1)
v_3 = new_array(v_1, v_2, n)
print(v_3)  # Output: [3, 1, 2, 7]
```

### Ex 3: Elementi comuni a due array

Scrivi una funzione che prende in input due array e la loro lunghezza e restituisce un array contente gli elementi comuni ad entrambi

	function elementi_comuni(A: integer[], B: integer[], n: integer, m: integer) -> integer[]:
		variable C: integer[]
		for i in [0, n) do:
			for j in [0, m) do:
				if A[i] == B[j] then:
					C.append(A[i])
				end if
			end for
		end for
		return C
	end function

```python
# Implementazione in Python [Assumiamo che gli elementi in A e B siano rispettivamente unici]
def elementi_comuni(A: list, B: list, n: int, m: int) -> list:
	C = []
	for i in range(n):
		for j in range(m):
			if A[i] == B[j]:
				C.append(A[i])
	return C
```

```python
# Esempio
A = [1, 2, 3, 4]
B = [3, 4, 5, 6, 7]
n = len(A)
m = len(B)
C = elementi_comuni(A, B, n, m)
print(C)  # Output: [3, 4]
```

### Ex 4: rimuovi le cifre dispari da un numero

Es: 1234 -> 24

	function rimuovi_cifre_dispari(n: integer) -> integer:
		variable r: integer
		variable i: integer
		variable cifra: integer
		r <- 0
		i <- 0
		while n > 0 do:
			cifra <- n mod 10
			if cifra mod 2 == 0 then
				r <- r + cifra*10^i
				i = i + 1
			end if
			n = n // 10
		end while
		return r
	end function

```python
# Implementazione in Python
def rimuovi_cifre_dispari(n: int) -> int:
    r = 0
    i = 0
    while n > 0:
        c = n % 10
        if c % 2 == 0:
            r = r + c * 10**i
            i = i + 1
        n = n // 10
    return r 
```

```python
# Esempio
a = rimuovi_cifre_dispari(123456)
print(a)
```

### Ex 5: Triangolo di Pascal

Scrivi una funzione che prende in input un numero intero e ritorna una lista di interi corrispondente l'n-esima riga del triangolo di Pascal. Il triangolo ha la seguente struttura:
    
1
   
1 1
  
1 2 1
 
1 3 3 1

1 4 6 4 1

MEMO: Puoi utilizzare il metodo .append().

	variable A: integer[]
	A <- [1, 2, 3]
	A.append(3)		# A = [1, 2, 3, 3]

	function triangolo_pascal(n: integer) -> integer[]:
		variable r_0: integer[]
		r_0 <- [1]
		for i in [0, n) do:
			variable r: integer[]
			r.append(1)
			for j in [0, i) do:
				r.append(A[i]+A[i+1])
			end for
			r.append(1)
			r_0 <- r
		end for
	end function

```python
# Implementazione in Python
def triangolo_pascal(n: int) -> list:
    r_0 = [1]
    for i in range(n):
        r = []
        r.append(1)
        for j in range(i):
            r.append(r_0[j]+r_0[j+1])
        r.append(1)
        r_0 = r
    return r_0
```

```python
triangolo_pascal(7)
```
